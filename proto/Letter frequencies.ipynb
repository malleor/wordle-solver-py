```python
import pandas as pd
import string
```

```python
with open('../res/valid_guesses.csv', 'rt') as f:
    d = f.readlines()
d = [w[:-1] for w in d]
```

```python
stat = pd.DataFrame(0, index=list(string.ascii_lowercase), columns=[0,1,2,3,4])
for w in d:
    for i in range(5):
        stat.loc[w[i]][i] += 1
```

```python
stat
```

```python
scores = [(w, sum([stat.loc[w[i]][i] for i in range(5)])) for w in d]
scores = sorted(scores, key=lambda x: -x[1])
```

```python
print('Top 100 words:')
for w, s in scores[:100]:
    print('  ', w, '', s)
```
