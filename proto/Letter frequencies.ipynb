```python
import pandas as pd
import string
```

```python
with open('../res/valid_guesses.csv', 'rt') as f:
    d = f.readlines()
with open('../res/valid_solutions.csv', 'rt') as f:
    d += f.readlines()
d = [w[:-1] for w in d]
```

```python
stat = pd.DataFrame(0, index=list(string.ascii_lowercase), columns=[0,1,2,3,4])
for w in d:
    for i in range(5):
        stat.loc[w[i]][i] += 1
stat /= len(d)
```

```python
stat
```

```python
scores = [(w, sum([stat.loc[w[i]][i] for i in range(5)])) for w in d]
scores = sorted(scores, key=lambda x: -x[1])
```

```python
print('Top 100 words:')
for w, s in scores[:100]:
    print('  ', w, '', s)
```

```python
stats = {l: [0]*5 for l in string.ascii_lowercase}
for w in d:
    for i, c in enumerate(w):
        stats[c][i] += 1
stats
```

```python
for w in d[::1000]:
    s = [stats[c][i] for i, c in enumerate(w)]
    print(w, s, sum(s)/5, max(s))
```

```python
word_scores = {w: sum([stats[c][i] for i, c in enumerate(w)]) for w in d}
```

```python
selection = d[::1000]
sorted([(w, word_scores[w]) for w in selection], key=lambda x: -x[1])[0][0]
```
